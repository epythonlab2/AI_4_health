#### **1.1 Measures of Central Tendency:**
**Central tendency** refers to the "typical" or "middle" value of a dataset. In ML, these are essential for understanding feature distributions and handling missing values (**imputation**).

**1. Mean (Average):** The sum of all values divided by the number of values.

**Formula**: $\bar{x} = \frac{1}{n} \sum_{i=1}^{n} x_i$

**Usecase:** Best for symmetric data without extreme outliers.

**Data Example: Test Scores**

The Scenario: Let's say we have test scores: `85, 90, 75, 92, 88`

**Q:** What is the center of the test scores?

**Step-by-Step Calculation (Python Implementation)**

```python
# Dataset
test_scores = [85, 90, 75, 92, 88]
# Find the n(Length of the dataset)
n = len(test_scores)
# Step 1: Calculate the sum()
total_sum = sum(test_scores)
# Step 2: Find the mean by divide total_sum by n
mean = total_sum / n

print(f"Dataset: {test_scores}")
print(f"Step 1 Sum: {total_sum}")
print(f"Step 2 Mean: {mean}")
```

```python
# Alternative using numpy
import numpy as np
# Calculations
mean_val = np.mean(test_scores)
print(f"Mean(Average): {mean_val}")
```

**Key Insight:** $86$ represents the "center" of our data. If we were building a model to predict future test scores without any other information, the mean would be your safest initial guess.

**2. Median (Middle Value)**: The middle value after sorting the data from least to greatest.

**Use Case:** Highly robust to outliers (e.g., house prices or salaries).

**Q:** What is the middle value of the above `test_score` data?

**Interactive Implementation**

```python
# Dataset: test scores
test_scores1 = [85, 90, 75, 92, 88]
test_scores2 = [85, 90, 75, 92, 88, 80]

# Calculations
median_val1 = np.median(test_scores1)

print(f"Dataset 1: {test_scores1}")

median_val2 = np.median(test_scores2)

print(f"Dataset 2: {test_scores2}")
print("-" * 30)
print(f"Median (Middle 1): {median_val1}")
print(f"Median (Middle 2): {median_val2}")
```

#### **1.2 Measures of Spread (Dispersion)**

While measures of central tendency show where the center of the data lies, **measures of spread** describe how widely the data is distributed. In **ML**, this helps us detect outliers and decide whether feature scaling is necessary.

**1. Variance** ($\sigma^2$) measures the average squared distance from the mean.

- **Why square it?** It ensures all distances are positive and penalizes extreme outliers more heavily.

- **Formula:**  $\sigma^2 = \frac{1}{n} \sum_{i=1}^{n} (x_i - \mu)^2$

**2. Standard Deviation ($\sigma$)** is the square root of the variance. It tells us how far data points typically are from the mean.

- **Why use it?** It is in the same units as your original data (e.g., if data is in "dollars," standard deviation is in "dollars," whereas variance is in "dollars squared").
- **Formula**: $\sigma = \sqrt{\sigma^2}$

- **Intuition:**

  - **Small **  data points are close to the mean

  - **Large **  data points are more spread out

**3. Range and Interquartile Range (IQR)**

- **Range:** The range measures the total spread of the data. It is difference between the highest and lowest values ($Max - Min$).
  - **Formula:** $Range = Max - Min$

- **Interquartile Range (IQR)** The IQR measures the spread of the **middle 50% of the data**. It is calculated as the difference between the third quartile and the first quartile.
  - **Formula:** $IQR = Q3 - Q1$


  - **Why IQR matters** IQR is robust to outliers, making it especially useful when data contains extreme values.

```python
import numpy as np

# Sample Data: Test scores
data = [85, 90, 75, 92, 88]

# 1. Range
data_range = np.max(data) - np.min(data)

# 2. Variance
variance = np.var(data)

# 3. Standard Deviation
std_dev = np.std(data)

# 4. IQR
q75, q25 = np.percentile(data, [75 ,25])
iqr = q75 - q25

print(f"Dataset: {data}")
print("-" * 30)
print(f"Range: {data_range}")
print(f"Variance: {variance:.2f}")
print(f"Standard Deviation: {std_dev:.2f}")
print(f"IQR: {iqr:.2f}")
```

**Key Insights:**
- The standard deviation (5.97) shows that scores typically vary by about 6 points around the mean, indicating moderate spread.

- The IQR (5.0) confirms that the middle 50% of the data is tightly grouped, so most observations are close to the mean.

- The range (17) captures the full spread but is influenced by the lowest score (75).

**ML Insight:** The dataset is well-centered with limited dispersion and no strong outliers, making it suitable for Z-score normalization and reliable model training.

**Box Plot:** A box plot is a simple visual summary of a dataset that shows its center, spread, and outliers using the median, quartiles, and range.

```python
import matplotlib.pyplot as plt

data = [85, 90, 75, 92, 88]

plt.figure(figsize=(8, 4))
plt.boxplot(data, vert=False, patch_artist=True, boxprops=dict(facecolor='#27ae60', alpha=0.5))
plt.title("Visualizing the Spread: Test Scores")
plt.xlabel("Scores")
plt.grid(linestyle='--', alpha=0.7)
plt.show()
```

**Observations:**
- **Box (IQR):** The green box shows the Interquartile Range from 85 (Q1) to 90 (Q3). This means the middle 50% of scores fall within a narrow 5-point range.

- **Median:** The line(**Orange**) inside the box marks the median (88). Its position slightly toward the right indicates a mild left skew.

- **Whiskers(Range):** The whiskers show the typical minimum and maximum values. The upper whisker extends to 92.

- **Outlier(Circle):** The point at 75 is flagged as a potential outlier, standing apart from the rest of the scores.

#### **1.3 Shape of Data**

While the mean and spread describe where data is and how wide it is, the shape describes the overall form of the distribution, including its symmetry and tails.

**1. Skewness: Is the data leaning?**

Skewness measures how asymmetric a dataset is.

- **Positive Skew (Right-skewed):** The "tail" extends toward higher values. The mean is usually greater than the median (e.g., Household income).

- **Negative Skew (Left-skewed):** The "tail" extends toward lower values. The mean is usually less than the median (e.g., Age of retirement).

- **Symmetric Distribution:** Left and right sides are roughly mirror images. The mean  median. Example: Human height

**2. Kurtosis: How "heavy" are the tails?**
Kurtosis describes the "fatness" of the tails and the sharpness of the peak.

- **High Kurtosis (Leptokurtic):** Sharp peak and heavy tails. This indicates a higher likelihood of extreme outliers.

- **Low Kurtosis (Platykurtic):** Flat peak and thin tails. Data is more spread out and less likely to have extreme outliers.

**Shape Analysis in Python**
We use the `scipy.stats` library to quantify these shapes. Let's use the Sleep Health dataset context (Sleep Duration).

**Scenario:**
- Identify where most sleep hours are concentrated

- Determine whether the distribution is skewed (left or right)

- Observe how outliers influence the shape of the data

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis

# Sample data: Sleep hours of 10 people (includes short sleepers and long sleepers)
sleep_hours = [7, 7.5, 8, 7.2, 6.8, 7.1, 7.4, 9.5, 10, 5]

# Calculate skewness and kurtosis
data_skew = skew(sleep_hours)
data_kurt = kurtosis(sleep_hours)

# Plot the distribution
plt.figure(figsize=(10, 5))
sns.histplot(sleep_hours, kde=True, color='skyblue')

# Add mean and median for interpretation
plt.axvline(np.mean(sleep_hours), color='red', linestyle='--', label='Mean')
plt.axvline(np.median(sleep_hours), color='green', linestyle='-', label='Median')

plt.title(f"Sleep Distribution\nSkew: {data_skew:.2f}, Kurtosis: {data_kurt:.2f}")
plt.xlabel("Hours of Sleep")
plt.legend()
plt.show()

```

**Interpretation:**

**Histogram & Density Curve:**

- The bars show how many people sleep a certain number of hours.

- The smooth line shows the overall pattern of sleep hours.

**Mean and Median:**

- The average sleep (mean, red dashed line) is about 7.4 hours.

- The middle value (median, green line) is around 7.2 hours.

- Since the mean is slightly higher than the median, it means a few people sleep much more than most  causing a small right skew.

**Skewness & Kurtosis:**

- **Skew:** 0.21  Slightly right-skewed, but close to symmetric.

- **Kurtosis:** -0.03  Very close to 0, indicating a distribution similar to a normal bell curve in terms of tail heaviness.

**Outliers:**

- Very low (5 hours) or very high (10 hours) sleep amounts stand out as unusual values. These are mild outliers.

Overall, the data is roughly symmetric, with a small positive skew due to higher sleep outliers. The mean and median being close together reinforces that the distribution isnt heavily skewed

```python
import matplotlib.pyplot as plt
import seaborn as sns

# -----------------------------
# Visualization Utilities
# -----------------------------

def plot_histograms(data, columns, bins=15, figsize=(6, 4), kde=True):
    """Plot histograms for selected numeric columns."""
    for col in columns:
        plt.figure(figsize=figsize)
        sns.histplot(data[col], bins=bins, kde=kde)
        plt.title(f'Distribution of {col}')
        plt.xlabel(col)
        plt.ylabel('Count')
        plt.tight_layout()
        plt.show()


def plot_boxplots(data, columns, figsize=(6, 4)):
    """Plot boxplots to visualize outliers."""
    for col in columns:
        plt.figure(figsize=figsize)
        sns.boxplot(x=data[col])
        plt.title(f'Boxplot of {col}')
        plt.tight_layout()
        plt.show()


def plot_distribution_by_target(data, feature_cols, target_col, bins=15, figsize=(6, 4)):
    """Compare feature distributions grouped by a target variable."""
    for col in feature_cols:
        plt.figure(figsize=figsize)
        sns.histplot(
            data=data,
            x=col,
            hue=target_col,
            bins=bins,
            kde=True,
            multiple="stack"
        )
        plt.title(f'{col} Distribution by {target_col}')
        plt.xlabel(col)
        plt.ylabel('Count')
        plt.tight_layout()
        plt.show()
```

**Part 5: Outliers**

- Identify potential outliers using boxplots or IQR method.

- Observation: Identifying Extreme Values and Data Errors

```python
# -----------------------------
# Part 5: Outliers (IQR Method)
# -----------------------------
for col in ['Glucose', 'Insulin', 'BMI', 'BloodPressure']:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5*IQR
    upper_bound = Q3 + 1.5*IQR
    outliers = data[(data[col] < lower_bound) | (data[col] > upper_bound)]
    print(f"\n{col} outliers:\n", outliers[[col]])
```

Upon examining outliers in Glucose, Insulin, and BMI, they can be classified into two categories:

**1. Physiologically Impossible Values (Data Errors)**
In health datasets, a value of 0 often indicates missing or unrecorded information.

**Where:**

  - 1: Success(Disease Present)

  - In our dataset, $p$ represents the proportion of diabetic patients and is approximately:
    - $p \approx 0.349$ .
- $1-p$ is the probability of failure, also called the complement of $p$($x=0$), often denoted as ($q$).
  - In our data:
    -  $1-p \approx 0.651$.

  How the Formula Calculates Outcomes:
  
  **Case 1: Patient has Diabetes** ($x = 1$)$$P(X=1) = p^1(1-p)^{1-1} = p \cdot (1-p)^0 = \mathbf{p}$$The exponent $(1-x)$ becomes 0, cancelling out the "failure" side of the equation and leaving only the probability of disease.
  
  **Case 2: Patient is Healthy** ($x = 0$)$$P(X=0) = p^0(1-p)^{1-0} = 1 \cdot (1-p)^1 = \mathbf{1-p}$$The exponent $(x)$ becomes 0, cancelling out the "success" side and leaving only the probability of the complement.

**2. Binomial Distribution**

The Binomial distribution models the number of successes ($k$) in a fixed number of independent Bernoulli trials ($n$), where each trial has a constant probability of success ($p$).

To find the probability of exactly $k$ successes in $n$ trials:
$$P(X=k) = \binom{n}{k} p^k (1-p)^{n-k}$$
**Where:**
- $\binom{n}{k}$: The "Binomial Coefficient" (calculated as $\frac{n!}{k!(n-k)!}$), which tells us how many different ways we can arrange $k$ successes in $n$ trials.
- $p^k$: The probability of getting $k$ successes.
- $(1-p)^{n-k}$: The probability of getting the remaining failures.

**Real Problem Example:** **Clinic Screening**

Suppose you are managing a small clinic. Based on your dataset, the probability of a patient having diabetes is $p = 0.349$. If 10 patients ($n=10$) walk in today, what is the probability that exactly 3 ($k=3$) of them have diabetes?

**1. Calculate the Combinations:** $\binom{10}{3} = 120$ ways to have 3 sick patients out of 10.

**2. Successes:** $0.349^3 \approx 0.0425$

**3. Failures:** $(1-0.349)^7 \approx 0.0494$4.

**Final Result:** $120 \times 0.0425 \times 0.0494 \approx \mathbf{0.252}$

**Interpretation:** There is a **25.2% chance** that exactly 3 out of the 10 patients will be diabetic.

**Why this matters for ML:**

The Binomial distribution is essential when you move into **Evaluation Metrics**. For example, if your model predicts diabetes for a group of patients, you can use this distribution to see if the number of "true positives" it found is statistically expected or just a result of random chance.

Is also  know as the **Normal Distribution**. Unlike the **Binomial distribution** which deals with "jumps" (3 patients vs 4 patients), continuous distributions deal with measurements that can have infinite decimals.

**Key Characteristics**
- **Symmetry**: The left side is a mirror image of the right.
- **Central Tendency:** The Mean, Median, and Mode are all at the exact center.
- **The Empirical Rule (68-95-99.7)**:
  - 68% of data falls within 1 Standard Deviation ($\sigma$) of the mean.
  - 95% of data falls within 2 Standard Deviations ($\sigma$).
  - 99.7% of data falls within 3 Standard Deviations ($\sigma$).

- $\mu$: Mean (the center of the peak).

###**3. Hypothesis Testing & Confidence Intervals**

####**3.1 What Is Hypothesis Testing?**

It helps us decide whether an observed effect is **real** or **due to chance**.

**For Example:** a difference in Glucose levels between diabetic and non-diabetic patientsis real or due to random chance.



####**3.2 Hypothesis Structure**

Every test starts with two competing statements:
- **Null Hypothesis ($H_0$):** The "Status Quo." It assumes there is no effect or no difference.
- **Alternative Hypothesis ($H_1$):** The "Claim." It assumes an effect exists.
- **Example from our Dataset:**
  - **$H_0$:** There is no difference in the mean BMI of diabetic vs. non-diabetic patients.
  - **Alternative Hypothesis ($H_1$ or $H_a$):** The claim you are trying to prove. It suggests that an effect or difference actually exists.

####**3.3 Test Statistic (Z-score)**

To decide between $H_0$ and $H_1$, we calculate a Test Statistic. The $Z$-score is a common choice when our sample size is large and we know the population variance.

**Explanation:**
- It measures how many standard deviations the sample mean is from the expected (null) mean.
- It is scaled by variability and sample size, meaning more data or less "messy" data makes the $Z$-score more reliable.

**The $Z$-score Formula:**$$Z = \frac{\bar{x} - \mu}{\sigma / \sqrt{n}}$$
- $\bar{x}$: Sample Mean
- $\mu$: Population Mean (under $H_0$)
- $\sigma$: Standard Deviation
- $n$: Sample Size

**Interpretation:**
- A large absolute value of the Z-score or t-score means the observed result is unlikely to be due to random chance under($H_0%$).

- Values close to zero indicate the sample is consistent with the null hypothesis.

####**3.4 P-value**

The P-value represents the probability of observing results at least as extreme as those obtained, assuming the null hypothesis is true.

**Decision rule:**
- **Low P-value (11$\le 0.05$):** The result is "**statistically significant**." We reject $H_0$.
- **High P-value ($> 0.05$):** We "fail to reject" $H_0$. There isn't enough evidence to support the change.

####**3.5 Confidence Intervals (The "Safety Margin")**

While hypothesis testing gives a "**Yes/No**" result, a **Confidence Interval** provides a range of values.

- **Meaning:** If we calculate a **95%** Confidence Interval for the mean Glucose level, we are saying we are **95%** confident the "true" population average falls within that specific range.

####**3.6 Real-World Example: Diabetes Testing**

Using your health dataset context, let's look at a practical application:

- **Scenario:** You observe that diabetic patients in your data have a higher average BMI than non-diabetic patients.
- **The Question:** Is this BMI difference a true medical pattern, or did we just happen to pick a few high-BMI people for this specific study?
- **The Test:**
  - $H_0$**: The average BMI is the same for both groups.
  - **$H_1$:** The average BMI is significantly higher in the diabetic group.

**Result:** If the test yields a P-value of 0.001, we reject the Null Hypothesis. We are now statistically confident that BMI is linked to diabetes in the broader population.

**Hypothesis Test: Glucose vs. Outcome**

We will test if patients with diabetes (Outcome 1) have higher glucose levels than those without (Outcome 0).
- Null Hypothesis ($H_0$): There is no difference in mean glucose between the two groups.

- Alternative Hypothesis ($H_1$): Diabetic patients have a higher mean glucose level.

```python
import numpy as np
import scipy.stats as stats

df = data.copy()

# 1. Split the data into two groups (removing 0 values identified in Q-Q plot)
# Using 'Glucose_Fixed' if you already handled the 0s, or filtering here:
group_0 = df[(df['Outcome'] == 0) & (df['Glucose'] > 0)]['Glucose']
group_1 = df[(df['Outcome'] == 1) & (df['Glucose'] > 0)]['Glucose']

# 2. Perform Independent T-Test
t_stat, p_val = stats.ttest_ind(group_1, group_0)

# 3. Calculate 95% Confidence Interval for the DIFFERENCE between means
# This tells us how much higher diabetic glucose is than non-diabetic
mean_diff = group_1.mean() - group_0.mean()
# Standard Error of the difference
se_diff = np.sqrt(stats.sem(group_1)**2 + stats.sem(group_0)**2)
# Degrees of freedom (approximate)
df_total = len(group_1) + len(group_0) - 2

ci_low, ci_high = stats.t.interval(0.95, df_total, loc=mean_diff, scale=se_diff)

# 4. Print Results
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_val:.10f}") # Use p_val (defined in step 2)
print(f"95% CI for the Difference: [{ci_low:.2f}, {ci_high:.2f}]")

# 5. Decision Logic
alpha = 0.05
if p_val < alpha:
    print(f"Decision: Reject the Null Hypothesis (P < {alpha})")
    print(f"Conclusion: Diabetic patients have significantly higher glucose (Avg diff: {mean_diff:.2f})")
else:
    print("Decision: Fail to Reject the Null Hypothesis")
```

**2. Handling Outliers**

Since the medically impossible zeros have been removed and imputed, our data is much "cleaner". However, we still have statistical **outliersvalues** that are physically possible but extremely rare (e.g., a BMI of 67, Insuline 864, or a Glucose of 199). These can disproportionately pull the mean and inflate the variance in our **T-tests and ML models**.

Here is how to mathematically identify and handle these extreme values.

**1. Detection: The Interquartile Range (IQR) Method** is the industry standard for identifying outliers. It focuses on the "middle 50%" of your data to define what is "normal".

image.png

**The Calculation Logic:**
- **Q1 (25th Percentile):** The value below which 25% of the data falls.
- **Q3 (75th Percentile):** The value below which 75% of the data falls.
- **IQR:** The distance between them ($Q3 - Q1$).
- **Lower Bound:** $Q1 - 1.5 \times \text{IQR}$
- **Upper Bound:** $Q3 + 1.5 \times \text{IQR}$

Anything outside these bounds is statistically considered an outlier.

**Implementation: Identifying Outliers**

You can run this loop to see exactly how many extreme values remain in each column after the zero-imputation step.

```python
def detect_outliers(data, columns):
    """
    Analyzes and reports outliers using the Interquartile Range (IQR) method.
    """
    print(f"{'Feature':<25} | {'Outliers':<10} | {'Lower Bound':<12} | {'Upper Bound':<12}")
    print("-" * 68)

    for col in columns:
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1

        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Count values outside the bounds
        outlier_count = ((data[col] < lower_bound) | (data[col] > upper_bound)).sum()

        # Print formatted row
        print(f"{col:<25} | {outlier_count:<10} | {lower_bound:<12.2f} | {upper_bound:<12.2f}")

# --- Execution ---
cols_to_check = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness',
                 'Insulin', 'BMI', 'DiabetesPedigree', 'Age']

detect_outliers(df_clean, cols_to_check)
```

**1. Log Transformation**

Used for right-skewed features with long tails. It compresses large values and spreads out smaller ones, making the distribution more symmetric.

**Formula**

$$x' = \log(x + c)$$
: original value

: small constant (often 1) to handle zeros

x: transformed value

**Why the $+c$ is critical**

The constant $c$ (usually $1$) is used because $\log(0)$ is undefined ($-\infty$). By using $x+1$, you ensure that:
- Zeros remain zeros: $\log(0 + 1) = 0$.
- All values stay positive: As long as your data is non-negative, the result is non-negative.

```python
# Log transform skewed features (handles zeros safely)

log_cols = ["Insulin", "DiabetesPedigree"]
df_clean.loc[:, log_cols] = np.log1p(df_clean[log_cols])
```

**Why it helps**

- Reduces the impact of extreme values

- Stabilizes variance

- Makes patterns easier for models to learn

Commonly applied to features like Insulin or Income, where a few large values dominate the scale.

**2. Robust Scaling**

**Robust scaler** is a feature scaling method that uses robust statistics instead of the mean and standard deviation. Unlike **StandardScaler**, it is not **sensitive** **to outliers.**

It is best when you want to keep all values while reducing the influence of outliers, especially for distance-based models.

**How it works**

**RobustScaler** centers data using the **median** and scales it using the interquartile range **(IQR)**

**Formula:**
$$x' = \frac{x - \text{median}(x)}{Q_3 - Q_1}$$


**Where:**
    $$IQR(x) = Q3 - Q1$$

- Q1: 25th percentile

- 3: 75th percentile

**Why it helps**

- Numerator $(x - \text{median})$: Centers the data around zero.
- Denominator $(Q_3 - Q_1)$: Scales the data based on the spread of the "middle 50%" of data values.
- Outliers have minimal impact compared to standard scaling

**When to use it**

- Data has strong outliers

- You are using distance-based models like KNN, SVM, K-Means

- After log or power transformations

Commonly used before models like **KNN, SVM,** and **linear models** when outliers are present.


```python
from sklearn.preprocessing import RobustScaler

# 1. Initialize scaler
scaler = RobustScaler()
# 2. Define features for specific scaling
robust_features = ['BloodPressure', 'SkinThickness', 'BMI', 'Insulin', 'Dian']
# 2. Separate features and target
features = df_clean.drop('Outcome', axis=1)
outcome = df_clean['Outcome'] # Keep the processed outcome

# 3. Scale
scaled_features = scaler.fit_transform(features)

# 4. Convert back to DataFrame AND keep the original index
df_scaled = pd.DataFrame(scaled_features, columns=features.columns, index=df_clean.index)

# 5. Re-attach the outcome
df_scaled['Outcome'] = outcome
```

**For example,** a **blood pressure** value of **72** (close to the median) is scaled to **0.00**, while a lower value such as **40** is mapped to approximately **2.00**.

This compression reflects how far a value lies from the typical range of the data, not its absolute difference.

As a result, extreme measurements are **down-weighted**, while clinically normal values remain centered around zero.

###**5. Correlation Analysis: Choosing the Right Lens**

Correlation is a statistical technique used to determine if a relationship exists between two variables and how strong that relationship is.
- It measures the strength and direction of linear relationships between two quantitative variables..

####**5.1 Pearson Correlation Coefficient ($r$)**: Measures linear relationships.

 - **$r = 1$:** Perfect positive correlation (as $X$ goes up, $Y$ goes up proportionally).
 - **$r = -1$:** Perfect negative correlation (as $X$ goes up, $Y$ goes down).
 - **$r = 0$:** No linear relationship exists.

 **Directionality:** A positive $r$ means variables move together; a negative $r$ means they move in opposite directions.

image.png

####**5.2 Spearman Rank Correlation ($\rho$)**

Measures the monotonic relationship between two variables.
- **Logic:** Converts raw values into ranks (1st, 2nd, 3rd, ) and then measures how well those ranks move together.

- **Robustness:** Insensitive to scale and far less affected by outliers since it depends only on order, not exact values.


image.png

####**5.3 Correlation Heatmap**
**A correlation heatmap** visually summarizes the relationships between all feature variables in a single grid.

- Each cell shows the correlation coefficient between two variables

- Color intensity represents the strength and direction of the relationship

- Makes patterns, strong associations, and potential multicollinearity easy to spot at a glance


```python
def plot_correlation_heatmap(df, size=(8, 6), method='spearman'):
    """
    Generates a correlation heatmap for a given DataFrame.

    Parameters:
    df (pd.DataFrame): The dataframe to analyze.
    title (str): The title of the plot.
    size (tuple): The width and height of the figure.
    """
    # 1. Compute the correlation matrix
    corr_matrix = df.corr(method=method)

    # 2. Initialize the matplotlib figure
    plt.figure(figsize=size)

    # 3. Create a mask to hide the upper triangle (optional but cleaner)
    # mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

    # 4. Draw the heatmap
    sns.heatmap(corr_matrix,
                annot=True,
                fmt=".2f",
                cmap='coolwarm',
                linewidths=0.5,
                square=True,
                center=0) # Ensures 0 is the neutral color

    plt.title("Diabetic Features Correlation Heatmap")
    plt.show()

# To use it:
plot_correlation_heatmap(df_scaled, method='pearson')
```

####**5.6 Regression**

**"Can I predict the outcome?**"Regression goes a step further. It creates a mathematical model (a "line of best fit") to predict a Dependent Variable (y) based on one or more Independent Variables (x).

The simplest form is Simple Linear Regression:$$y = \beta_0 + \beta_1x + \epsilon$$

**where:**

-  = dependent (target) variable

-  = independent (predictor) variable

- $_0$ = intercept (value of  when  = 0)

- $_1$ = slope (change in for a one-unit change in )

-  = error term (captures variation not explained by the model)

**In words:** the model predicts  as a linear function of x, plus some random error.


image.png

**1. The Blue Points ($y$):** These are the actual patient data (e.g., $x$ = Age, $y$ = Blood Pressure). In the real world, data never falls perfectly on a line.

**2. The Red Line ($y = \beta_0 + \beta_1x$):** This is the models prediction, showing the underlying trend in the data.

- **Intercept ($\beta_0$):** This is the point where the red line hits the vertical axis ($x=0$). In medicine, this might represent the "baseline" value. For example, if $x$ is "years of smoking," $\beta_0$ would be the expected lung capacity of a non-smoker.

- **Slope ($\beta_1$):** This is the "Steepness." It tells you how much $y$ changes for every 1-unit increase in $x$. A steeper slope means $x$ is a very strong predictor of $y$.

**3. The Purple Line / Error ($\epsilon$):** This is the Residual. It is the distance between the actual patient data point and the prediction line.
- If $\epsilon$ is small, your model is very accurate.
- If $\epsilon$ is large, there are other factors (noise) affecting the patient that the model isn't capturing.

**Clinical Use Case:** **Dose-Response Prediction**

**Research Question:** Can we predict a patients Blood Sugar Reduction ($y$) based on their Medication Dosage ($x$)?

```python
from sklearn.linear_model import LinearRegression

# 1. Prepare data
X = df_clean[['Age']]
y = df_clean['Glucose']

# 2. Fit the model
model = LinearRegression()
model.fit(X, y)

beta_0 = model.intercept_
beta_1 = model.coef_[0]

# 3. Visualization
plt.figure(figsize=(10, 6))

# Use sns.regplot to automatically calculate and plot the Confidence Interval
# ci=95 is the default; it shows the 95% confidence interval for the regression estimate
sns.regplot(x='Age', y='Glucose', data=df_clean,
            scatter_kws={'alpha':0.5},
            line_kws={'color':'red', 'label':f"Slope (1): {beta_1:.4f}"},
            ci=95)

# Intercept line
plt.axhline(y=beta_0, color='green', linestyle='--', label=f"Intercept (0): {beta_0:.2f}")

# Labels and title
plt.title("Clinical Linear Regression with 95% Confidence Interval")
plt.xlabel("Age")
plt.ylabel("Glucose Level")
plt.legend()
plt.grid(True, linestyle=':', alpha=0.6)
plt.show()

# Print slope info
print(f"For every 1 unit increase in Age, Glucose increases by {beta_1:.2f} units.")
```

  - **Slope ( = 0.6834):** Glucose increases by ~0.68 mg/dL for each unit increase in scaled age, showing a mild upward trend.
